```python
import os.path 
import numpy as np
from PIL import Image

# Folder where the created images will be saved in
save_path = r'/local/data1/elech646/Tumor_grade_classification/resized_single_channel_datasets/resized_t2_t1ce_flair'

HGG_path = save_path + "/HGG"
LGG_path = save_path + "/LGG"

# Create subfolders
G4 = save_path + "/G4"
if not os.path.exists(G4):
    os.mkdir(G4)
    
G3 = save_path + "/G3"
if not os.path.exists(G3):
    os.mkdir(G3)    
    
G2 = save_path + "/G2"
if not os.path.exists(G2):
    os.mkdir(G2)
    
filepath = []
# Scanning files + directories
for roots, dirs, files in os.walk("/local/data1/elech646/Tumor_grade_classification/single_channel_slices/slices_t2_t1ce_flair"):
    for name in files:
        if name.endswith(".png"):
            filepath.append(roots + os.path.sep + name)
            
G2_patients = ['TCGA-CS-4944', 'TCGA-CS-6667', 'TCGA-CS-6668', 'TCGA-CS-6669',
              'TCGA-DU-5849', 'TCGA-DU-5853', 'TCGA-DU-5871', 'TCGA-DU-5872',
              'TCGA-DU-5874', 'TCGA-DU-6407', 'TCGA-DU-7008', 'TCGA-DU-7014',
              'TCGA-DU-7015', 'TCGA-DU-7294', 'TCGA-DU-7301', 'TCGA-DU-7306',
              'TCGA-DU-8164', 'TCGA-DU-8166', 'TCGA-DU-8167', 'TCGA-DU-A5TR',
              'TCGA-DU-A5TS', 'TCGA-DU-A5TU', 'TCGA-DU-A6S2', 'TCGA-DU-A6S3',
              'TCGA-FG-5964', 'TCGA-FG-6689', 'TCGA-FG-6690', 'TCGA-FG-6691',
              'TCGA-FG-7634', 'TCGA-FG-8189', 'TCGA-FG-A4MT', 'TCGA-FG-A713',
              'TCGA-HT-7473', 'TCGA-HT-7602', 'TCGA-HT-7604', 'TCGA-HT-7605',
              'TCGA-HT-7608', 'TCGA-HT-7680', 'TCGA-HT-7692', 'TCGA-HT-7693',
              'TCGA-HT-7884', 'TCGA-HT-8018', 'TCGA-HT-8107', 'TCGA-HT-8113',
              'TCGA-HT-A61A']
    
# Loop through the directories 
for file in filepath:
    filename = file.split("/")[-1].split(".png", 1)[0]
    if filename.startswith("BraTS20"):
        folder_name = HGG_path
        path = file.split("HGG/")[-1].split("/")[0] + "/"
        plane = file.split("HGG/")[0].split('/slices_t2_t1ce_flair')[-1]
        patient_name = "_".join(filename.split("_")[0:3]) + "/"
        grade = G4
                
        if not os.path.exists(G4 + plane):
            os.mkdir(G4 + plane)

        if not os.path.exists(G4 + plane + path):
            os.mkdir(G4 + plane + path)
    else:
        folder_name = LGG_path
        path = file.split("LGG/")[-1].split("/")[0] + "/"
        plane = file.split("LGG/")[0].split('/slices_t2_t1ce_flair')[-1]
        patient_name = filename.split("_")[0] + "/" 
        p_name = filename.split("_")[0]

        
        G2_cond = 'LGG' in folder_name and p_name in G2_patients
        
        if G2_cond:
            grade = G2
        else:
            grade = G3

        if G2_cond and not os.path.exists(G2 + plane): #+ patient_name):
            os.mkdir(G2 + plane) #+ patient_name)
        elif not G2_cond and not os.path.exists(G3 + plane):
            os.mkdir(G3 + plane)
        
        if G2_cond and not os.path.exists(G2 + plane + path[:-1]): #+ patient_name):
            os.mkdir(G2 + plane + path[:-1]) #+ patient_name)
        elif not G2_cond and not os.path.exists(G3 + plane + path[:-1]):
            os.mkdir(G3 + plane + path[:-1])

    # Open images
    img = Image.open(file)
    
    # Set desired width + height
    new_size = (224, 224)
    
    # Resize images
    resized_image = img.resize(new_size)
    
    # Extract file name
    if 'LGG' in folder_name:
        name = grade + plane + path #+ patient_name
    else:
        name = grade + plane + patient_name
    
    # Rename
    title = name + filename + '.png'
    
    # Save resized images      
    resized_image.save(os.path.join(save_path, title))
    #print(os.path.join(save_path, title))
```
